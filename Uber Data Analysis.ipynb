## Dim_DataTime

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
df.info()
```

```python
Dim_DataTime = df[['tpep_pickup_datetime','tpep_dropoff_datetime']].drop_duplicates().reset_index(drop=True)
```

```python
Dim_DataTime['pick_hour'] = Dim_DataTime['tpep_pickup_datetime'].dt.hour
Dim_DataTime['pick_day'] = Dim_DataTime['tpep_pickup_datetime'].dt.day
Dim_DataTime['pick_month'] = Dim_DataTime['tpep_pickup_datetime'].dt.month
Dim_DataTime['pick_year'] = Dim_DataTime['tpep_pickup_datetime'].dt.year
Dim_DataTime['pick_weekday'] = Dim_DataTime['tpep_pickup_datetime'].dt.weekday
```

```python
Dim_DataTime['drop_hour'] = Dim_DataTime['tpep_dropoff_datetime'].dt.hour
Dim_DataTime['drop_day'] = Dim_DataTime['tpep_dropoff_datetime'].dt.day
Dim_DataTime['drop_month'] = Dim_DataTime['tpep_dropoff_datetime'].dt.month
Dim_DataTime['drop_year'] = Dim_DataTime['tpep_dropoff_datetime'].dt.year
Dim_DataTime['drop_weekday'] = Dim_DataTime['tpep_dropoff_datetime'].dt.weekday
```

```python
Dim_DataTime['DataTime_ID'] = Dim_DataTime.index
```

```python
Dim_DataTime
```

```python
#reorder columns
Dim_DataTime = Dim_DataTime[['DataTime_ID', 'tpep_pickup_datetime', 'pick_hour', 'pick_day', 'pick_month', 'pick_year', 'pick_weekday',
                             'tpep_dropoff_datetime', 'drop_hour', 'drop_day', 'drop_month', 'drop_year', 'drop_weekday']]
Dim_DataTime
```

## Dim_Rate_Code

```python
# Mapping of rate codes to names
rate_code_names = {
    1: 'Standard rate',
    2: 'JFK',
    3: 'Newark',
    4: 'Nassau or Westchester',
    5: 'Negotiated fare',
    6: 'Group ride'
}

# Create a new column 'Rate_code_name' in the existing DataFrame
df['Rate_code_name'] = df['RatecodeID'].map(rate_code_names)
# Create a new DataFrame 'Dim_Rate_Code' with selected columns
Dim_Rate_Code = df[['RatecodeID', 'Rate_code_name']]
print(Dim_Rate_Code)
```

```python
Dim_Rate_Code = df[['RatecodeID','Rate_code_name']].drop_duplicates().reset_index(drop=True)
```

```python
Dim_Rate_Code['Rate_code_ID'] = Dim_Rate_Code.index
Dim_Rate_Code
```

```python
Dim_Rate_Code = Dim_Rate_Code[['Rate_code_ID','RatecodeID','Rate_code_name']]
```

## Dim_Payment_Type

```python
# Mapping of payment type to names
payment_type_name = {
    1: 'Credit card',
    2: 'Cash',
    3: 'No charge',
    4: 'Dispute',
    5: 'Unknown',
    6: 'Voided trip'
}

# Create a new column 'payment_type_name' in the existing DataFrame
df['payment_type_name'] = df['payment_type'].map(payment_type_name)
# Create a new DataFrame 'Dim_Payment_Type' with selected columns
Dim_Payment_Type = df[['payment_type', 'payment_type_name']]
print(Dim_Payment_Type)
```

```python
# Filter the DataFrame to include only rows with payment_type 5 or 6
filtered_df = Dim_Payment_Type[(Dim_Payment_Type['payment_type'] == 5) | (Dim_Payment_Type['payment_type'] == 6)]
print(filtered_df)
```

```python
Dim_Payment_Type = df[['payment_type', 'payment_type_name']].drop_duplicates().reset_index(drop=True)
```

```python
Dim_Payment_Type['Payment_Type_ID'] = Dim_Payment_Type.index
Dim_Payment_Type
```

```python
Dim_Payment_Type = Dim_Payment_Type[['Payment_Type_ID','payment_type', 'payment_type_name']]
```

## Fact_Table

```python
Fact_Table_Merged = pd.merge(Fact_Table_Merged, Dim_Dropoff_Location[['Dropoff_Location_ID','dropoff_longitude','dropoff_latitude']], on=['dropoff_longitude','dropoff_latitude'])

```

```python
Fact_Table_Merged.shape
```

```python
Fact_Table_Merged = pd.merge(Fact_Table_Merged, Dim_Pickup_Location[['Pickup_Location_ID', 'pickup_longitude','pickup_latitude']], on=['pickup_longitude','pickup_latitude'])
Fact_Table_Merged.shape 
```

```python
Fact_Table_Merged = pd.merge(Fact_Table_Merged, Dim_DataTime[['DataTime_ID', 'tpep_pickup_datetime','tpep_dropoff_datetime']], on=['tpep_pickup_datetime','tpep_dropoff_datetime'])
Fact_Table_Merged.shape 
```

## Connection with MS SQL

```python
# Replace 'database_name' with your actual database name
database_name = 'Uber_OLAP'

# Windows authentication connection string for MS SQL Server
conn_str = (
    r'Driver={SQL Server};'
    r'Server=.;'
    f'Database={database_name};'
    r'Trusted_Connection=yes;'
)
```

```python
# Create a connection to the database
conn = pyodbc.connect(conn_str)
engine = create_engine('mssql+pyodbc://', creator=lambda: conn)
```

```python
# Create tables in MS SQL Server 
Dim_DataTime.to_sql('Dim_DataTime', engine, if_exists='replace', index=False)
Dim_Passenger_Count.to_sql('Dim_Passenger_Count', engine, if_exists='replace', index=False)
Dim_Trip_Distance.to_sql('Dim_Trip_Distance', engine, if_exists='replace', index=False)
Dim_Rate_Code.to_sql('Dim_Rate_Code', engine, if_exists='replace', index=False)
Dim_Payment_Type.to_sql('Dim_Payment_Type', engine, if_exists='replace', index=False)
Dim_Dropoff_Location.to_sql('Dim_Dropoff_Location', engine, if_exists='replace', index=False)
Dim_Pickup_Location.to_sql('Dim_Pickup_Location', engine, if_exists='replace', index=False)
Fact_Table_Merged.to_sql('Fact_Table_Merged', engine, if_exists='replace', index=False)
```

```python
# Close the connection
conn.close()
```
